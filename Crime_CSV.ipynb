```python
import pandas as pd
import json
import os
import csv
##
file = "_Change_Notice__Police_Department_Incidents.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
len(file_df)
```

```python
file_df = file_df.sort_values(["Date"], ascending=False)
file_df
```

```python
# Get a reference to the column names
file_df.columns
```

```python
# Rename column headers
file_df = file_df.rename(columns={"IncidntNum":"ID",
                                       "Category":"Category",
                                       "Descript":"Descript",
                                       "DayOfWeek":"Day Of Week",
                                       "Date":"Date",
                                       "Time":"Time",
                                       "PdDistrict":"District",
                                       "Resolution":"Resolution"})
file_df.columns
```

```python
# Extract some columns
file_df = file_df.loc[:,['Category', 'Descript', 'Day Of Week', 'Date', 'Time', 'District',
       'Resolution', 'Address', 'X', 'Y']]
file_df.head()
```

```python
# Create a new columns for Year, Day and Month
file_df["Year"] = file_df["Date"].str.rstrip('/').str.split('/').str[2]
file_df["Day"] = file_df["Date"].str.rstrip('/').str.split('/').str[1]
file_df["Month"] = file_df["Date"].str.rstrip('/').str.split('/').str[0]
file_df = file_df.sort_values(["Year"], ascending=False)
file_df
```

```python
file_df["Year"].value_counts()
```

```python
# Select 2016, 2017 and 2018
Years = ["2016", "2017", "2018"]
reduced_file_df = file_df[file_df["Year"].isin(Years)]



reduced_file_df.head()
```
